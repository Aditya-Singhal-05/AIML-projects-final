```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

# ---------------------------------
# Step 1: Create dataset in code
# ---------------------------------
data = [
    [8.5, 80, 85, 78, 75, 82, 88, 40, 90, 35, 4, 2, 3, 1, 0, 1, 0, "AI Engineer"],
    [7.8, 75, 80, 72, 70, 76, 85, 35, 88, 30, 3, 1, 2, 1, 0, 1, 0, "AI Engineer"],
    [8.0, 78, 82, 75, 72, 80, 84, 45, 86, 40, 3, 1, 2, 1, 0, 1, 0, "Data Scientist"],
    [7.5, 72, 75, 70, 68, 74, 78, 40, 80, 35, 2, 1, 1, 1, 0, 1, 0, "Data Scientist"],
    [7.2, 65, 78, 60, 75, 70, 72, 88, 35, 30, 3, 1, 2, 0, 1, 0, 0, "Web Developer"],
    [7.0, 60, 74, 58, 78, 68, 70, 90, 30, 28, 2, 1, 1, 0, 1, 0, 0, "Web Developer"],
    [8.1, 76, 88, 74, 70, 79, 90, 55, 45, 40, 4, 2, 2, 0, 0, 0, 0, "Software Engineer"],
    [7.9, 74, 85, 72, 68, 77, 87, 50, 40, 35, 3, 1, 2, 0, 0, 0, 0, "Software Engineer"],
    [8.3, 78, 76, 70, 72, 80, 74, 35, 30, 90, 3, 2, 2, 0, 0, 0, 1, "Cyber Security Analyst"],
    [8.0, 75, 74, 68, 70, 78, 72, 30, 28, 88, 2, 1, 1, 0, 0, 0, 1, "Cyber Security Analyst"],
    [8.7, 82, 87, 80, 78, 84, 89, 42, 92, 38, 5, 2, 4, 1, 0, 1, 0, "AI Engineer"],
    [7.6, 70, 77, 69, 74, 73, 75, 85, 38, 32, 3, 1, 1, 0, 1, 0, 0, "Web Developer"],
    [8.2, 79, 83, 77, 73, 81, 86, 43, 84, 37, 4, 2, 3, 1, 0, 1, 0, "Data Scientist"],
    [8.4, 77, 86, 75, 71, 80, 88, 52, 46, 42, 4, 2, 2, 0, 0, 0, 0, "Software Engineer"],
    [8.1, 76, 75, 71, 72, 79, 73, 34, 32, 91, 3, 2, 2, 0, 0, 0, 1, "Cyber Security Analyst"]
]

columns = [
    "CGPA", "Aptitude", "Programming", "Maths", "Communication",
    "Logical_Reasoning", "DSA", "Web_Dev", "AI_ML", "Cyber_Security",
    "Projects", "Internships", "Certifications",
    "Interest_AI", "Interest_Web", "Interest_Data", "Interest_Security",
    "Career_Label"
]

df = pd.DataFrame(data, columns=columns)

# ---------------------------------
# Step 2: Encode target labels
# ---------------------------------
label_encoder = LabelEncoder()
df["Career_Label"] = label_encoder.fit_transform(df["Career_Label"])

# ---------------------------------
# Step 3: Split features and target
# ---------------------------------
X = df.drop("Career_Label", axis=1)
y = df["Career_Label"]

# ---------------------------------
# Step 4: Train model
# ---------------------------------
model = RandomForestClassifier(n_estimators=200, random_state=42)
model.fit(X, y)

# ---------------------------------
# Step 5: Take user input
# ---------------------------------
print("Enter student details:")

cgpa = float(input("CGPA: "))
aptitude = int(input("Aptitude Score (0-100): "))
programming = int(input("Programming Score (0-100): "))
maths = int(input("Maths Score (0-100): "))
communication = int(input("Communication Score (0-100): "))
logical_reasoning = int(input("Logical Reasoning Score (0-100): "))
dsa = int(input("DSA Score (0-100): "))
web_dev = int(input("Web Development Score (0-100): "))
ai_ml = int(input("AI/ML Score (0-100): "))
cyber_security = int(input("Cyber Security Score (0-100): "))
projects = int(input("Number of Projects: "))
internships = int(input("Number of Internships: "))
certifications = int(input("Number of Certifications: "))
interest_ai = int(input("Interest in AI (1 for Yes, 0 for No): "))
interest_web = int(input("Interest in Web Dev (1 for Yes, 0 for No): "))
interest_data = int(input("Interest in Data Science (1 for Yes, 0 for No): "))
interest_security = int(input("Interest in Cyber Security (1 for Yes, 0 for No): "))

# ---------------------------------
# Step 6: Create proper user input DataFrame
# ---------------------------------
user_data = pd.DataFrame([[
    cgpa, aptitude, programming, maths, communication,
    logical_reasoning, dsa, web_dev, ai_ml, cyber_security,
    projects, internships, certifications,
    interest_ai, interest_web, interest_data, interest_security
]], columns=X.columns)

# ---------------------------------
# Step 7: Predict career
# ---------------------------------
predicted_label = model.predict(user_data)
predicted_career = label_encoder.inverse_transform(predicted_label)

print("\nRecommended Career Path:", predicted_career[0])

# ---------------------------------
# Step 8: Show confidence score
# ---------------------------------
probabilities = model.predict_proba(user_data)
confidence = max(probabilities[0]) * 100
print("Prediction Confidence:", round(confidence, 2), "%")

# ---------------------------------
# Step 9: Skill gap analysis
# ---------------------------------
def suggest_skills(aptitude, programming, dsa, web_dev, ai_ml, cyber_security, communication):
    suggestions = []

    if programming < 70:
        suggestions.append("Improve Programming")
    if dsa < 70:
        suggestions.append("Practice Data Structures and Algorithms")
    if web_dev < 70:
        suggestions.append("Learn Web Development")
    if ai_ml < 70:
        suggestions.append("Learn AI and Machine Learning basics")
    if cyber_security < 70:
        suggestions.append("Build Cyber Security fundamentals")
    if communication < 70:
        suggestions.append("Improve Communication Skills")
    if aptitude < 70:
        suggestions.append("Practice Aptitude and Logical Reasoning")

    return suggestions

skills_needed = suggest_skills(
    aptitude, programming, dsa, web_dev, ai_ml, cyber_security, communication
)

print("\nSkill Gap Analysis:")
if len(skills_needed) == 0:
    print("You are doing well in all major skill areas.")
else:
    for skill in skills_needed:
        print("-", skill)
```
